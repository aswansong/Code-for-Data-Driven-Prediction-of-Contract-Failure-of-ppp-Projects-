```python
from fpdf import FPDF
```

```python
class Detected_Image:
    
    """
    image_size = [width, height], 
    image: path or URL of the image (the image with boxes)
    
    """
    def __init__(self, image, image_name, image_size, defect_type):
        self.image = image
        self.image_name = image_name
        self.image_size = image_size
        self.defect_type = defect_type
        
    @property
    def Image(self):
        return self.image
    @property
    def Image_Name(self):
        return self.image_name
    @property
    def Image_Size(self):
        return self.image_size
    @property
    def Defect_Type(self):
        return self.defect_type
```

```python
def new_report(unit='mm', format='A4'):
    new_pdf = FPDF(unit='mm', format='A4')
    return new_pdf

"""
Insert the content to the pdf in function 'new_report',
image: Detected_Image object
new_pdf: the pdf in the function 'new_report',
scale: the scale used to zoom the image, the default one is 34.0

"""

def add_content(image, new_pdf, scale=34.0):
    #add new page, set the font format and the frame of the page
    new_pdf.add_page()
    new_pdf.set_font('Arial', style='', size=10)
    new_pdf.rect(8.0, 8.0, 194.0, 282.0)
     
    new_pdf.set_xy(x=8.0, y=8.0)
    new_pdf.cell(w=30, h=6, 
                 txt='Name', border=1, 
                 ln=0, align='center')
    new_pdf.cell(w=99, h=6, 
                 txt=image.Image_Name, border=1, 
                 ln=0, align='center')
    new_pdf.cell(w=30, h=6, 
                 txt='Size', border=1, 
                 ln=0, align='center')
    new_pdf.cell(w=35, h=6, 
                 txt=f'{image.Image_Size[0]}Ã—{image.Image_Size[1]}', border=1, 
                 ln=0, align='center')

    new_pdf.set_xy(x=8.0, y=14.0)
    new_pdf.cell(w=30, h=6, 
                 txt='Defect type', border=1, 
                 ln=0, align='center')
    new_pdf.cell(w=99, h=6, 
                 txt=image.Defect_Type, border=1, 
                 ln=0, align='center')
    new_pdf.cell(w=30, h=6, 
                 txt='PE name', border=1, 
                 ln=0, align='center')
    new_pdf.cell(w=35, h=6, 
                 txt='', border=1, 
                 ln=0, align='center')

    new_pdf.set_xy(x=8.0, y=20.0)
    new_pdf.cell(w=30, h=6, 
                 txt='Comments', border=1, 
                 ln=0, align='center')
    new_pdf.cell(w=164, h=6, 
                 txt='', border=1, 
                 ln=0, align='center')
        
    new_pdf.image(image.Image, 30.0, 28.0, 
                  image.Image_Size[0]/scale, image.Image_Size[1]/scale, 
                  link='', type='')
    
    return pdf       

def save_report(pdf, location):
    pdf.output(location, 'F')
```

```python
pdf = new_report()
image_size = [5184, 3888]
image = Detected_Image('C:\\Users\\Meghdad\\Downloads\\results\\DJI_20200702101212_0005_SUPR.JPG_msk.png',
                       'DJI_20200702101212_0005_SUPR.JPG_msk.png', 
                       image_size,
                       'crack')
# can add more image here.

report = add_content(image=image, 
                     new_pdf=pdf)
save_report(report, 'C:\\Users\\Meghdad\\Desktop\\test.pdf')
```
