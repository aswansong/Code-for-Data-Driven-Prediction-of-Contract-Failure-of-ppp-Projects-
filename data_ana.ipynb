```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
pd.options.display.max_rows = 10
pd.options.display.max_columns = None
```

```python
df1  = pd.read_excel('1990-2000.xlsx')
df2 = pd.read_excel('2001-2011.xlsx')
df3 = pd.read_excel('2012-2022.xlsx')
df_all = pd.concat([df1,df2,df3],axis=0,ignore_index=True)
select_df = df_all[['Region','ContractPeriod','Country','IncomeGroup',
        'IDA Status','Type of PPI','Project name','Project status',
        'Financial closure year','Capacity',
        'TotalInvestment','PhysicalAssets','Sponsors',
        'Sponsors Country','BiLateralSupport','Primary sector',
        'UnsolicitedProposal','PublicDisclosure','Subtype of PPI',
        'GovtGrantingContract','PercentPrivate','Subsector',
       'Segment','Main Revenue Source']]
select_df
```

```python
def trainsform_float(se):
    for i in range(len(se)):
        if  se[i] == 'Not Available' or se[i] =='Not Applicable':
            se[i]=0.00001
    se = se.astype(float)
    se_mean = round(se.mean())
    print(se_mean)
    for i in range(len(se)):
        if se[i]==0.00001:
            se[i]=se_mean
    return se
    
select_df['ContractPeriod'] = trainsform_float(select_df['ContractPeriod'])
select_df['Capacity'] = trainsform_float(select_df['Capacity'])
select_df['PercentPrivate'] = trainsform_float(select_df['PercentPrivate'])
select_df['PhysicalAssets'] = trainsform_float(select_df['PhysicalAssets'])
select_df['TotalInvestment'] = trainsform_float(select_df['TotalInvestment'])
select_df
```

```python
def sponser_number(se):
    for i in range(len(se)):
        try:
            se[i] = se[i].count(',')+1
        except:
            pass
    se = se.astype(float)
    return se
def border_country(se1,se2):
    for i in range(len(se1)):
        try:
            l = se1[i].split(', \n\r')
            # print(l)
            if se2[i] not in l or len(l)>1:
                se1[i] = 1
            else:
                se1[i]=0
        except:
            pass
    return se1
def income_level(se):
    for i in range(len(se)):
        if se[i]=='Low income':
            se[i]=0
        elif se[i] == 'Upper middle income':
            se[i] = 2
        else:
            se[i]=1
    return se
def Type_of_PPI(se):
    for i in range(len(se)):
        if se[i]=='Greenfield project':
            se[i] = 1
        elif se[i]=='Brownfield' or se[i]=='Divestiture' or se[i]=='Management and lease contract':
            se[i] = 0
    return se
def Project_status(se):
    for i in range(len(se)):
        if se[i]=='Active' or se[i]=='Concluded':
            se[i]=0
        elif se[i]=='Cancelled' or se[i]=='Distressed':
            se[i]=1
    return se
        
select_df['Sponsors'] = sponser_number(select_df['Sponsors'])
select_df['Sponsors Country'] = border_country(select_df['Sponsors Country'],select_df['Country'])
select_df['IncomeGroup'] = income_level(select_df['IncomeGroup'])
select_df['Type of PPI'] = Type_of_PPI(select_df['Type of PPI'])
select_df['Project status'] = Project_status(select_df['Project status'])
select_df
```

```python
def yes_no(se):
    for i in range(len(se)):
        if se[i]=='Yes':
            se[i]=0
        elif se[i]=='No':
            se[i]=1
    return se
def IDA_Status(se):
    for i in range(len(se)):
        if se[i]=='Blended':
            se[i]=1
        elif se[i]=='IDA':
            se[i]=0
        elif se[i]=='Non-IDA':
            se[i]=2
    return se

select_df['UnsolicitedProposal'] = yes_no(select_df['UnsolicitedProposal'])
select_df['PublicDisclosure'] = yes_no(select_df['PublicDisclosure'])
select_df['BiLateralSupport'] = yes_no(select_df['BiLateralSupport'])
select_df['IDA Status'] = IDA_Status(select_df['IDA Status'])
select_df
```

```python
def wgi_data_add(sheetname):
    wgi_data = pd.read_excel('wgidataset.xlsx',header=[13,14],sheet_name=sheetname)
    for index,row in wgi_data.iteritems():
        # print(index)
        if row.dtype=="float64":
            mean_value = row.mean()
            row.fillna(mean_value,inplace=True)
    country_wgi = wgi_data[('Unnamed: 0_level_0', 'Country/Territory')]
    add_list = []
    for i in range(len(select_df['Country'])):
        try:
            index_number = country_wgi[country_wgi.values == select_df['Country'][i]].index
            if select_df['Financial closure year'][i]%2==0:
                a = select_df['Financial closure year'][i]
                add_list.append(wgi_data.loc[index_number[0],(a, 'Estimate')])
            elif select_df['Financial closure year'][i]%2==1:
                a = select_df['Financial closure year'][i]-1
                add_list.append(wgi_data.loc[index_number[0],(a, 'Estimate')])
        except:
            add_list.append(0)
    select_df.insert(select_df.shape[1], sheetname, add_list)
wgi_data_add('VoiceandAccountability')
wgi_data_add('Political StabilityNoViolence')
wgi_data_add('GovernmentEffectiveness')
wgi_data_add('RegulatoryQuality')
wgi_data_add('RuleofLaw')
wgi_data_add('ControlofCorruption')
select_df
```

```python
wdi_data = pd.read_csv('WDI/data.csv')
data_by = wdi_data.groupby('Series Name')
for index_progect,df_wdi in data_by:
    new_df = df_wdi.copy(deep=True)
    new_df.reset_index(drop=True, inplace=True)
    for name,row in new_df.iteritems():
        if 'YR' in name:
            for i in range(len(row)):
                if row[i]=='..':
                    row[i] =0.000001
            row = row.astype(float)
            mean_wid  = round(row.mean())
            for i in range(len(row)):
                if row[i]==0.000001:
                    row[i]=mean_wid
                    
    add_list =[]
    country_name  = new_df['Country Name']
    for i in range(len(select_df['Country'])):
        # index_number = country_name[country_name.values == select_df['Country'][i]].index
        # a = select_df['Financial closure year'][i]
        # if len(index_number)!=0:
        #     add_list.append(new_df.loc[index_number[0],f'{a} [YR{a}]'])
        try:
            index_number = country_name[country_name.values == select_df['Country'][i]].index
            a = select_df['Financial closure year'][i]
            if len(index_number)!=0:
                add_list.append(new_df.loc[index_number[0],f'{a} [YR{a}]'])
            else:
                mean_value = new_df[f'{a} [YR{a}]'].mean()
                # print(mean_value)
                add_list.append(mean_value)
        except:
            add_list.append(0)
    # print(add_list)
    try:
        select_df.insert(select_df.shape[1], index_progect, add_list)
    except:
        pass
select_df
```
